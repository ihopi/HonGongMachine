#### 선형 회귀 관련 자료

```python
import numpy as np
import matplotlib.pyplot as plt

# sklearn 설치할 때 반드시 'pip install scikit-learn' 형태로 입력 해 주어야 한다.
# 애매한 install을 제거하기 위해서 scikit 설치 형태는 지원하지 않기로 하였음
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
# 훈련 세트와 테스트 세트 나눔
train_input, test_input, train_target, test_target = train_test_split(
perch_length, perch_weight, random_state=42)
```

```python
# 훈련 세트와 테스트 세트 2차원 배열로 변경
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
```

```python
knr = KNeighborsRegressor(n_neighbors=3)

# k-최근접 이웃 회귀 모델을 훈련합니다
knr.fit(train_input, train_target)
```

```python
print(knr.predict([[50]]))
```

```python
distances, indexes = knr.kneighbors([[50]])
plt.scatter(train_input, train_target)
# 훈련 세트 중에서 이웃 샘플만 다시 그립니다
plt.scatter(train_input[indexes], train_target[indexes], marker='D')
# 50cm 농어 데이터
plt.scatter(50, 1033, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

#### 최근접으로 처리하는 경우, 위 처럼 정확하지 않은 무게가 산출될 수 있다.
<br>
그러므로, 가장 중요한 내용 중 하나는 선형예측을 (선형회귀) 하기 위해서 그래프를 그리는 방식에 있다
아래 자료를 확인 해 보면, 50 의 길이이거나 100의 길이이거나 정상적으로 무게가 예측되지 않는다.

```python
print(np.mean(train_target[indexes]))
```

```python
distances, indexes = knr.kneighbors([[100]])
print(np.mean(train_target[indexes]))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# 선형 회귀 모델을 훈련합니다.
lr.fit(train_input, train_target)

# 50cm 농어에 대해 예측 해 봅니다.
print(lr.predict([[50]]))
```

```python
#선형회귀 훈련도 산점도를 그려본다.
plt.scatter(train_input, train_target)

#15 에서 50 Rkwl 1차 방정식 그래프를 그린다
plt.plot([15, 50], [15*lr.coef_+lr.intercept_, 50*lr.coef_+lr.intercept_])

# 50 cm 농어 데이터
plt.scatter(50, 1241.8, marker="^")
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
train_poly = np.column_stack((train_input ** 2, train_input))
test_poly = np.column_stack((test_input ** 2, test_input))
```

```python
print(train_poly.shape, test_poly.shape)
```

```python
sample_poly = train_poly[:5]
print(sample_poly)
```

```python
lr = LinearRegression()
lr.fit(train_poly, train_target)

print(lr.predict([[80**2, 80]]))
```

```python
print(lr.coef_, lr.intercept_)
```

```python
# 구간별 직선을 그리기 위해 15에서 49까지 정수 배열을 만든다
point = np.arange(15, 80)

# 훈련 세트의 산점도를 그립니다.
plt.scatter(train_input, train_target)

# 15 에서 49까지 2차 방정식 그래프를 그립니다
# 아래 내용은 ax**2 - bx + c 의 형태임
# 어떻게 이런 내용이 나올 수 있는지에 대해서 확인 해 보도록 하자.
plt.plot(point, 1.01*point**2 - 21.6*point + 116.05)

# 50cm 농어 데이터
plt.scatter(50, 1574, marker='^')
plt.scatter(80, 4883, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(lr.score(train_poly, train_target))
```

```python
print(lr.score(test_poly, test_target))
```

테스트 결과물이 더 높다. 그러므로 과소적합 상태가 되어 있으므로, 이를 해결하기 위해 3-3 특성공학 - 다중회귀에 대해서 알아보도록 한다
